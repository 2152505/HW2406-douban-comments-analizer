```python
for i in range(1, 10):
    for j in range(1, 10):
        result = i * j
        print(f"{i} x {j} = {result}")
    print()
```

```python
for i in range(1, 10):
    for j in range(1, i+1):
        result = i * j
        print(f"{i} x {j} = {result}\t", end="")
    print()
```

```python
for i in range(1, 10):
    for j in range(1, 10):
        if j >= i:
            result = i * j
            print(f"{i} x {j} = {result}\t", end="")
        else:
            print("\t", end="")
    print()
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

# 读取图像
image = imread("1.jpg")

# 显示原始图像
plt.imshow(image)
plt.show()

# 将非蓝色通道设置为0
image_copy = image.copy()
image_copy[:, :, 0] = 0  # 红色通道
image_copy[:, :, 1] = 0  # 绿色通道

# 显示修改后的图像
plt.imshow(image_copy)
plt.show()
```
