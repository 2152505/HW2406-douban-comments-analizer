```python
import gazpacho
doubanURL="https://book.douban.com/top250?start=0"
```

```python
# 把网页取下来
doubanHTML = gazpacho.get(doubanURL)
doubanHTML
```

```python
# 规范html结构
doubanTree = gazpacho.Soup(doubanHTML)
doubanTree
```

```python
# 查找图书名信息
AllBookNameInDiv = doubanTree.find("div", attrs={"class":"pl2"})
AllBookNameInDiv
```

```python
BookNames=[]
for i in range(0,25):
    booknow=AllBookNameInDiv[i].find("a").strip()
    BookNames.append(booknow)
```

```python
BookNames=[]
for i in range(0,25):
    booknow=AllBookNameInDiv[i].find("a").strip()
    BookNames.append(booknow)
print (BookNames)
```

```python
# 保存到csv: comma separated values
import csv
with open("Top250-1.csv",mode="wt",encoding="utf-8",newline="") as BAOAN:
    GUANLIYUAN = csv.writer(BAOAN)
    GUANLIYUAN.writerow(["图书名",])
    for name in BookNames:
        GUANLIYUAN.writerow([name,])
print("文件保存完毕！")
```

```python
replaceDict = {
    "（":"(",
    "[":"(",
    "【":"(",
    "）":")",
    "]":")",
    "】":")"
}
replaceTable = str.maketrans(replaceDict)
replaceTable
```

```python
"[美] 哈珀·李 / 高红梅 / 译林出版社 / 2012-9 / 32.00元".translate(replaceTable)
```

```python
NationAndAuthorTemplate = "({nation}){author}"
```

```python
parse.parse(NationAndAuthorTemplate,AllBookInfoInP[-3].text.translate(replaceTable).split("/")[0])
```

```python
Nations = []
Authors = []
for eachP in AllBookInfoInP:
    extractedInfo = eachP.text.split("/")[0].translate(replaceTable).strip()
    if extractedInfo.startswith("("):
        parsedText = parse.parse(NationAndAuthorTemplate, extractedInfo)
        Nations.append(parsedText["nation"].strip())
        Authors.append(parsedText["author"].strip())
    else:
        Nations.append("中国")
        Authors.append(extractedInfo)

print(Nations,Authors)
```
